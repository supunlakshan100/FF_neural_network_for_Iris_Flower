```python
import pandas as pd

dataset=pd.read_csv ('/content/drive/MyDrive/Colab Notebooks/Deep Learning/Projects/FFNN for Iris Flower/iris.csv').values
print(dataset,dataset.shape)
```

```python
data=dataset[:,0:4]
target=dataset[:,4]

print(data.shape,target.shape)
```

```python
print (data)
print(target)
```

###convert to categorical format (Categorical Encording)

```python
import tensorflow as tf
New_target=tf.keras.utils.to_categorical(target)
```

```python
print (target[:10])
```

```python
print (New_target[:10])
```

```python
from sklearn.model_selection import train_test_split
train_data,test_data,train_target,test_target=train_test_split(data, New_target,test_size=0.1)
```

```python
print(train_data,train_data.shape)
```

```python
print(test_data,test_data.shape)
```

### Build A Neural Network

```python
from keras.models import Sequential
from keras.layers import Dense

model= Sequential() # empty sequential type neural network
model.add(Dense(64,input_dim=4,activation="relu")) # 1st Hidden layer
model.add(Dense(32,activation="relu")) # 2st Hidden layer
model.add(Dense(3,activation="softmax")) # output layer

model.compile (loss="categorical_crossentropy", optimizer="Adamax",metrics =["accuracy"])
model.summary()
```

```python
from sklearn.utils.validation import Parameter
for layer in model.layers:
  print("============",layer.name,"=============")

  parameters=layer.get_weights()
  print(parameters)
```

###  Trainning the Model

```python
model.fit(train_data,train_target, epochs=1000, validation_split=0.1)
```

```python
from matplotlib import pyplot as plt

plt.plot(model.history.history['loss'],label='loss')
plt.plot(model.history.history['val_loss'],label='val_loss')
plt.legend()
plt.xlabel('#epoch')
plt.ylabel('loss')
plt.show
```

```python
plt.plot(model.history.history['accuracy'],label='accuracy')
plt.plot(model.history.history['val_accuracy'],label='val_accuracy')
plt.legend()
plt.xlabel('#epoch')
plt.ylabel('loss')
plt.show
```

```python
import numpy as np

test_data=np.array([5.3,2.3,4.7,2.1]).reshape(1,-1)

predicted_target=model.predict(test_data)
print(predicted_target)
```

```python
label=np.argmax(predicted_target)
print(label)
```
